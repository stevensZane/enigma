```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO

# Chargement des donnes
data = """ID_Testeur,Date_Test..."""  # Votre CSV complet ici
df = pd.read_csv(StringIO(data))

# Configuration du style
plt.style.use('seaborn')
plt.figure(figsize=(12, 8))

# Boxplot pour la distribution globale
plt.subplot(2, 1, 1)
sns.boxplot(x=df['Tps_Execution_Sec'], color='skyblue', width=0.4)
plt.title('Distribution des Temps d\'Excution (Toutes Commandes)', fontsize=14)
plt.xlabel('Temps (secondes)', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Histogramme comparatif Russite/chec
plt.subplot(2, 1, 2)
success = df[df['Tache_Accomplie'] == True]['Tps_Execution_Sec']
fail = df[df['Tache_Accomplie'] == False]['Tps_Execution_Sec']

bins = [0, 0.5, 1, 1.5, 2, 3, 5, 10]
plt.hist(success, bins=bins, alpha=0.7, label='Commandes russies', color='green')
plt.hist(fail, bins=bins, alpha=0.7, label='Commandes choues', color='red')

plt.title('Comparaison des Temps par Rsultat', fontsize=14)
plt.xlabel('Temps (secondes)', fontsize=12)
plt.ylabel('Nombre de Commandes', fontsize=12)
plt.legend()
plt.grid(linestyle='--', alpha=0.5)
plt.yscale('log')  # chelle logarithmique pour mieux visualiser

plt.tight_layout()
plt.show()
```

```python
# Calcul des statistiques
df = pd.read_csv("yup.csv")
stats = df['Tps_Execution_Sec'].describe(percentiles=[.25, .5, .75, .9])
top_slow = df.nlargest(5, 'Tps_Execution_Sec')[['Commande_Profere', 'Tps_Execution_Sec', 'Tache_Accomplie']]

# Affichage
print("Statistiques cls (en secondes):")
print(f"- Moyenne: {stats['mean']:.2f}s")
print(f"- Mdiane: {stats['50%']:.2f}s")
print(f"- 75% des commandes < {stats['75%']:.2f}s")
print(f"- Maximum: {stats['max']:.2f}s (Commande: {top_slow.iloc[0]['Commande_Profere']})")

print("\nTop 5 des commandes les plus lentes:")
print(top_slow.to_markdown(index=False))

# Graphique des tops lents
plt.figure(figsize=(10, 6))
bars = plt.barh(top_slow['Commande_Profere'], top_slow['Tps_Execution_Sec'], 
                color=['red' if not x else 'orange' for x in top_slow['Tache_Accomplie']])
plt.title('Top 5 des Commandes les Plus Longues', fontsize=14)
plt.xlabel('Temps (secondes)', fontsize=12)
plt.grid(axis='x', linestyle='--')

# Annotation
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y()+0.3, f'{width:.2f}s', ha='left', va='center')

plt.tight_layout()
plt.show()
```
